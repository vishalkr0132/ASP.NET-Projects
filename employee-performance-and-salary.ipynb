```python
#import numpy, pandas libraries
import numpy as np
import pandas as pd
```

```python
# Question 1: Load the dataset and display the first 10 rows'
df = pd.read_csv('/kaggle/input/employee-performance-and-salary-dataset/dataset.csv')
print(df.head(10))
```

```python
# Question 2: Display the column and it's data types
print(df.dtypes)
```

```python
# Question 3: Number of rows and columns
print(df.shape)
```

```python
# Question 4: Selecting Name, Age, and Salary columns
df[['Name','Age','Salary']]
```

```python
#Question 5: Selecting first 5 rows using iloc
print(df.iloc[:5])
```

```python
# Question 6: Selecting all the rows where department is sales by using loc method
df.loc[df['Department']=='Sales']
```

```python
# Question 7: Selecting rows where Employees Salary > 8000 and Status Active
df[(df['Salary'] > 8000) & (df['Status'] == 'Active')]
```

```python
# Question 8: Display the employees with > 5 Experience and performance Score >= 4
df[(df['Experience'] > 5) & (df['Performance Score'] >= 4)]
```

```python
# Question 9: Display every alternate rows by using iloc slicing.
print(df.iloc[::2])
```

```python
# Question 10: Display employees in sales with top 10% salary
salary_threshold = df['Salary'].quantile(0.9)
sales_with_10 = df[(df['Department'] == 'Sales') & (df['Salary'] >= salary_threshold)]
print(sales_with_10)
```

```python
# Question 11: Replace all 'Inactive' with 'Retired' in Status column
df['Status'] = df['Status'].replace('Inactive','Retired')
```

```python
# Qestion 12: Add Bonus column with 10% of salary
df['Bonus'] = df['Salary'] * 10 / 100
```

```python
# Question 13: Update Performance Score for IT department to 5 if NaN
df[df['Department'] == 'IT']['Department'].fillna(5)
```

```python
# Question 14: Modify Experience for employees with less than 2 years to 2
df.loc[df['Experience'] < 2, 'Experience'] = 2
```

```python
# Question 15: Normalize Salary column with the help of min,max
df['Normalized Salary'] = (df['Salary'] - df['Salary'].min()) / (df['Salary'].max() - df['Salary'].min())
```

```python
# Question 16: Replace all Location with 'Remote' for employees with Performance Score below 3
df.loc[df['Performance Score'] < 3, 'Location'] = 'Remote'
```

```python
# question 17: Create experience level column based on the experience
def Experience_Level(Experience):
    if Experience <= 2:
        return 'Junior'
    elif Experience <= 7:
        return 'Mid'
    else:
        return 'Senior'

df['Experience Level'] = df['Experience'].apply(Experience_Level)
```

```python
# Question 18: Add a new row with specific values using the append function

# The append() function is not supported in this version of Pandas (Pandas 2.0+), 
# so we use pd.concat() to add the new row to the DataFrame.

new_row = {'Name': 'Vishal Kumar', 'Age': 24, 'Salary': 20000, 'Department': 'IT', 'Experience': 0, 'Performance Score': 4, 'Status': 'Active'}
df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)

# df.loc[len(df)] = new_row #alternative method by using loc
```

```python
# Question 19: Delete the session column
session = df['Session'] #for further process Session column is needed so that i restored.
df.drop(columns=['Session'],inplace=True)
```

```python
# Question 20: Delete rows where Age > 60
df.drop(df[df['Age'] > 60].index, inplace=True)
```

```python
# Question 21: Add a new column 'Age Group'
def Age_Group(Age):
    if Age < 30:
        return 'Young'
    elif Age >= 30 and Age <60:
        return 'Middle-aged'
    else:
        return 'Senior'

df['Age Group'] = df['Age'].apply(Age_Group)
```

```python
# Question 22: Identify the columns with missing values using isna()
df.isna().sum()
```

```python
# Question 23: Fill missing values in 'Performance Score' with average
Average = df['Performance Score'].mean()
df['Performance Score'] = df['Performance Score'].fillna(Average).round(1)
```

```python
# Question 24: Drop rows with missing values
df.dropna(inplace=True)
```

```python
# Question 25: Fill missing values in 'Performance Score' with the median of each Department
Group_Department = df.groupby('Department')['Performance Score']
df['Performance Score'] = Group_Department.transform(lambda x: x.fillna(x.median()))
```

```python
# Question 26: Identify employees in "IT" with Age above department average
it_avg_age = df[df['Department'] == 'IT']['Age'].mean()
Above_avg_it = df[(df['Department'] == 'IT') & (df['Age'] > it_avg_age)]
print(Above_avg_it.head())
```

```python
# Question 27: Remove duplicate rows based on Name and Joining Date
df = df.drop_duplicates(subset=['Name', 'Joining Date'])
```

```python
# Question 28: Remove rows with inconsistent Experience and Age
df.drop(df[df['Experience'] > (df['Age'] - 18)].index, inplace=True)
```

```python
# Question 29: Sort DataFrame by Salary (descending)
df = df.sort_values(by='Salary', ascending=False)
print(df.head())
```

```python
# Question 30: Sort DataFrame by Department and Joining Date (ascending)
df.sort_values(by=['Department', 'Joining Date'])
```

```python
# Question 31: Rank employees by Salary within Department
df['Salary Rank'] = df.groupby('Department')['Salary'].rank(ascending=False).astype(int)
print(df.head())
```

```python
# # Question 32: Sort by Experience (descending) with tiebreaker on Performance Score (ascending)
df.sort_values(by=['Experience', 'Performance Score'], ascending=[False, True])
```

```python
# Question 33: Filter for employees in "HR" with Salary above 5000
df_filtered = df[(df['Department'] == 'HR') & (df['Salary'] > 5000)]
df.head()
```

```python
# Question 34: Filter out rows where Location is "Chicago"
df=df[df['Location'] != 'Chicago']
print(df)
```

```python
# Question 35: Delete rows where Performance Score is below 2
df = df.drop(df[df['Performance Score'] < 2].index)
# df = df[df['Performance Score'] >= 2] #alternative method
```

```python
# Question 36: Find average Salary for each Department
avg_salary_by_dept = df.groupby('Department')['Salary'].mean()
print(avg_salary_by_dept)
```

```python
# Question 37: Calculate total Salary by Status
total_salary_by_status = df.groupby('Status')['Salary'].sum()
print(total_salary_by_status)
```

```python
# Question 38: Find maximum Performance Score for each Location
max_perf_by_location = df.groupby('Location')['Performance Score'].max()
print(max_perf_by_location)
```

```python
# Question 39: Calculate total Salary paid by joining year
df['Year Joined'] = pd.to_datetime(df['Joining Date']).dt.year
total_salary_by_year = df.groupby('Year Joined')['Salary'].sum()
print(total_salary_by_year)
```

```python
# Question 40: Find department with highest average Performance Score
highest_avg_perf_dept = df.groupby('Department')['Performance Score'].mean().idxmax()
print(highest_avg_perf_dept)
```

```python
# Question 41: Count employees by Location and Session
emp_count_by_location_session = df.groupby(['Location', session]).size()
print(emp_count_by_location_session)
```

```python
# Question 42: Calculate statistics (sum, mean, std) of Salary by Status
stats_by_status = df.groupby('Status')['Salary'].agg(['sum', 'mean', 'std'])
print(stats_by_status)
```

```python
# Question 43: Create new 'Monthly Salary' column by dividing Salary by 12
# df['Monthly Salary'] = df['Salary'].apply(lambda x: x / 12)
df.loc[:, 'Monthly Salary'] = df['Salary'].apply(lambda x: x / 12)
```

```python
df.shape
```

```python
# Question 44: Convert all string values to uppercase

df = df.applymap(lambda x: x.upper() if isinstance(x, str) else x)
df
```

```python
# Question 45: Inner join with a second dataset (Department and Budget)
budget_df = pd.DataFrame({'Department': ['Sales', 'IT', 'HR', 'Marketing'], 
                          'Budget': [80000, 100000, 90000, 60000]
                         })
pd.merge(df,budget_df)
```

```python
# Question 46: Merge with dataset containing employee ID and Projects
projects_df = pd.DataFrame({
    'Employee ID': df['ID'],
    'Projects': np.random.choice(['Project A', 'Project B', 'Project C', 'Project D'],len(df['ID']))
})
df = pd.merge(df,projects_df,left_on='ID',right_on='Employee ID')
```

```python
# Question 47: Left join with Team Lead details
team_leads_df = pd.DataFrame({
    'Department': ['Sales', 'IT', 'HR', 'Marketing'],
    'Team Lead': ['Vishal', 'Ayush', 'Prajwal', 'Santoosh']
})

df = pd.merge(df,team_leads_df,how='left')
```

```python
# Question 48: Create a DataFrame with employee ID and Project Count, then merge it with the original dataset,Calculate the average Salary per project.
project_counts_df = df.groupby('ID')['Projects'].count().reset_index(name='Project Count')
merged_df = df.merge(project_counts_df, on='ID', how='left')
merged_df['Avg Salary per Project'] = merged_df['Salary'] / merged_df['Project Count']
```

```python
# Question 49: Calculate percentage increase in Salary compared to department average
avg_salary_by_dept = df.groupby('Department')['Salary'].mean()
df['Salary Increase %'] = ((df['Salary'] - df['Department'].map(avg_salary_by_dept)) / df['Department'].map(avg_salary_by_dept)) * 100
```
